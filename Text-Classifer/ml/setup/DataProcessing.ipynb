# Data Processing

This notebook is used for mapping the positive and negative sentiment values to 1 and 0 respectively. The changes are saved into a file named data.csv which will be used within the tfx pipeline.

[Kaggle Dataset](https://www.kaggle.com/lakshmi25npathi/imdb-dataset-of-50k-movie-reviews)

```python
import pandas as pd
```

```python
df = pd.read_csv("IMDB_Dataset.csv")
df
```

```python
sentiment = {'positive': 1, 'negative': 0}
```

```python
df.sentiment = [sentiment[item] for item in df.sentiment]
df
```

```python
df.to_csv('data.csv', index=False)
```
